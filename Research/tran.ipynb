```python
from pymystem3 import Mystem
m = Mystem()


def lemmas_all(data):
    df = data['Название операции']
    all_word = ['операция','рекламировать','котомесяц','котодень','котонеделя','песонеделя','песомесяц','песодень','лечение','защищать','избавлять','вакцина','стерилизация','препарат','реклама','анализ','корм','игрушка']
    for word in df.split(' '):
        lemmas = m.lemmatize(word)
        for w in lemmas:
            for k in all_word: 
                if k == w:
                    return k
#lemmas_toy(tran['Название операции'].head(5))
tran['lemmas'] = tran.apply(lemmas_all, axis=1)
tran.head(5)
```

```python
def del_all(data):
    df = data['lemmas']
    all_word = ['операция','рекламировать','котомесяц','котодень','котонеделя','песонеделя','песомесяц','песодень','лечение','защищать','избавлять','вакцина','стерилизация','препарат','реклама','анализ','корм','игрушка']
    if df in all_word:
        return None
    return data['Название операции']
tran['Название операции'] = tran.apply(del_all, axis=1)
tran.head(5)
```

```python
def lemmas_cor_rek1(data):
    df = data['Название операции']
    word = 'рекламировать'
    f = 'Promote'
    if df is not None:
        if f in df:
            return 'реклама'
        
    if data['lemmas'] is not None:
        if data['lemmas'] == word:
            return 'реклама'
    return data['lemmas']
```

```python
tran['lemmas'] = tran.apply(lemmas_cor_rek1, axis=1)
tran.head(5)
```

```python
def lemmas_cor_rek2(data):
    df = data['Название операции']
    word = 'корм'
    f = ['Canin', 'Delicate', 'Палочки', 'корма','food','ProPlan','Кормить','Баночка','наполнителя']
    if df is not None:
        for i in f:
            if i in df:
                return 'еда'
        
    if data['lemmas'] is not None:
        if data['lemmas'] == word:
            return 'еда'
    return data['lemmas']
```

```python
tran['lemmas'] = tran.apply(lemmas_cor_rek2, axis=1)
```

```python
def del_all3(data):
    df = data['Название операции']
    f = ['Canin', 'Delicate', 'Палочки', 'корма','Promote','food','ProPlan','Паучи','Кормить','Баночка','подушечки','Litter']
    if df is not None:
        for i in f:
            if i in df:
                return None
    return data['Название операции']
tran['Название операции'] = tran.apply(del_all3, axis=1)
```

```python
def lemmas_tul(data):
    df = data['Название операции']
    all_word = 'туалет'
    if df is not None:
        for word in df.split(' '):
            lemmas = m.lemmatize(word)
            for w in lemmas:
                if w == all_word: 
                    return all_word
    return data['lemmas']
tran['lemmas'] = tran.apply(lemmas_tul, axis=1)
def del_all_tul(data):
    df = data['lemmas']
    all_word = 'туалет'
    if df is not None:
        if df == all_word:
            return None
    return data['Название операции']
tran['Название операции'] = tran.apply(del_all_tul, axis=1)
```

```python
def lemmas_drug(data):
    df = data['Название операции']
    f = ['подарок', 'Попрыгун', 'ПёсоНабор','Уборка','Лоток','Ремонт','Машинка для стрижки','Fresh','Переноска','клетки','box','Kitty','КотоНабор']
    if df is not None:
        for i in f:
            if i in df:
                return 'другое'
    return data['lemmas']
tran['lemmas'] = tran.apply(lemmas_drug, axis=1)
def del_all_drug(data):
    df = data['Название операции']
    f = ['подарок', 'Попрыгун', 'ПёсоНабор','Уборка','Лоток','Ремонт','Машинка для стрижки','Fresh','Переноска','клетки','box','Kitty','КотоНабор']
    if df is not None:
        for i in f:
            if i in df:
                return None
    return data['Название операции']
tran['Название операции'] = tran.apply(del_all_drug, axis=1)
```

```python
def lemmas_lec(data):
    df = data['Название операции']
    f = 'Toy'
    if df is not None:
        if f not in df:
                return 'лечение'
    return data['lemmas']
tran['lemmas'] = tran.apply(lemmas_lec, axis=1)
def del_all_lec(data):
    df = data['Название операции']
    f = 'Toy'
    if df is not None:
        if f in df:
                return df
    return None
tran['Название операции'] = tran.apply(del_all_lec, axis=1)
```
